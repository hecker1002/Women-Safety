```python
# function to find outliers in a Dataset 

def find_outlier(col) :
    q1 = col.quantile(0.25) # first quatile 
    q3 = col.quantile(0.75) # third quantile 
    iqr = q3 -q1 
    lower_bound = q1 - 1.5 * iqr 
    upper_bound = q3 + 1.5 * iqr 

    outlier = (col < lower_bound) | (col > upper_bound)
    return outlier 
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train , y_test  = train_test_split(X , y , test_size= 0.2 )

# Use K-Fold Cross Vald 
```

```python
x_train.keys()
```

```python
from sklearn.ensemble import RandomForestRegressor

reg_rf = RandomForestRegressor(n_estimators = 100 )

reg_rf.fit(x_train , y_train) 

y_pred = reg_rf.predict(x_test)
```

```python

from sklearn.metrics import r2_score 

r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import AdaBoostRegressor

ada_reg = AdaBoostRegressor(n_estimators=100)

ada_reg.fit(x_train , y_train)

ypred_ada= ada_reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score 

r2_score(y_test, ypred_ada)
```
